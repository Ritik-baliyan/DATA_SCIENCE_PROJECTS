```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/DependerKumarSoni/Naive-Bayes/main/adult.csv")
```

```python
col_names=['Age','Workclass','Fnlwgt','Education','Education_num','Marital_Status','Occupation','Relationship',
           'Race','Sex','Capital_gain','Capital_loss','Hours_per_week','Native_country','Income']
          
df.columns=col_names

df.columns
```

```python
## Checking Null values

df.isna().sum()
```

```python
## Checking the unique values

for column in df.columns:
  print(column, df[column].unique())
```

```python
##Counts of '?' in the columns

print(df['Workclass'].value_counts())
print(df['Occupation'].value_counts())
print(df['Native_country'].value_counts())
```

```python
## replacing the '?' with null values so that we can change the null values

df['Workclass'].replace( ' ?', np.NaN, inplace=True)
df['Occupation'].replace( ' ?', np.NaN, inplace=True)
df['Native_country'].replace( ' ?', np.NaN, inplace=True)
```

```python
## Checking Null values 

df.isna().sum()
```

```python
## Filling null values with most common values

df['Workclass'].fillna( df['Workclass'].mode().iloc[0], inplace=True)
df['Occupation'].fillna( df['Occupation'].mode().iloc[0], inplace=True)
df['Native_country'].fillna( df['Native_country'].mode().iloc[0], inplace=True)

```

```python

df.isna().sum()
```

```python
df
```

```python
X=df.drop(['Age','Fnlwgt','Education_num','Marital_Status','Relationship','Race','Sex','Capital_gain','Capital_loss','Native_country','Income'],axis=1)
y=df['Income']
```

```python
X_encoded=pd.get_dummies(X)
```

```python
## Scaling the data

from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X_encoded)
```

```python
## Train_test_split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.31,random_state=42)
```

```python
#Using Naive Bayes

from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predicted))
```

```python
#Confusion Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)

print(cm)
```
